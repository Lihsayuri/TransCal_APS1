```python
import numpy as np
from math import *
import matplotlib.pyplot as plt


Ts = 100 # ¬∫C
Tinf = 25 # ¬∫C
rAleta = 12e-3 # m
L = 168e-3 # m
h = 15 # W/(m^2*K) 5 15
k = 154.808 # W/(m*K) 190 154.08
P = 2*pi*rAleta
Atr = pi*(rAleta**2)

# Fazer gr√°fico da temperatura da aleta por comprimento da aleta
```

1 - a) Ajuste o modelo e defina as condi√ß√µes de contorno para obter um gr√°fico da distribui√ß√£o de temperatura (T_anal√≠tica) ao longo do comprimento da aleta. Qual a temperatura na extremidade da aleta?

```python
# teta/ tetab = T(x) - Tinf/ Tb - Tinf

# T(x) = 75* (cosh m*(L-x) + (h/m*k)*senh m(L-x))/(cosh m*L + (h/m*k)*senh m*L )

import math
import numpy as np

m = sqrt((h*P)/(k*Atr))
listaTemp = []
lista_x = np.linspace(0,L, 10000)
for x in lista_x:
    num = (cosh(m*(L-x)) + (h/(m*k))*sinh(m*(L-x)))
    den = (cosh(m*L) + (h/(m*k))*sinh(m*L))
    temp = ((num/den)*(Ts - Tinf)) + 25
    listaTemp.append(temp)


plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(lista_x, listaTemp, color="darkcyan")
plt.title("Gr√°fico de T(x) por x")
plt.xlabel("Valores de x [de 0 at√© L em metros]")
plt.ylabel("Valores de T(x) em Celsius - caso A")
```

1 - b) Determine a taxa de transfer√™ncia de calor, a efici√™ncia e a efetividade para a geometria da aleta usada. 

```python
# taxa de transfer√™ncia de calor
M = (Ts- Tinf)*sqrt(h*P*k*Atr)
qa = M*((np.sinh(m*L) + (h/(m*k))*cosh(m*L))/(cosh(m*L) + (h/(m*k))*np.sinh(m*L)))

areaTotal = Atr+(2*pi*rAleta*L)
taxa_transf_calor = qa
eficiencia = qa/(h*areaTotal*(Ts - Tinf))
efetividade = qa/ (h*Atr*(Ts-Tinf))

print(f'√Årea total {areaTotal}')
print(f'A taxa de transfer√™ncia de calor √© {taxa_transf_calor}')
print(f'A efici√™ncia √© {eficiencia}')
print(f'A efetividade √© {efetividade}')

```

* Se um bast√£o de cobre com o mesmo di√¢metro fosse usado em lugar do bast√£o de alum√≠nio, qual seria a mudan√ßa na taxa de remo√ß√£o de calor atrav√©s do bast√£o? Considere kCu= 400 W/(m*K) e ùëòAl = 240 W/(m*K)



```python
kCu = 400 #W/(mK)
kAl = 240 #W/(mK)
h_ = 5
r = 0.05
L_ = 0.1

# se ambos tem o mesmo di√¢metro, tamb√©m possuem o mesmo raio, que ser√° chamado de r;
# Portanto, o per√≠metro do bast√£o circular ser√° o mesmo, j√° que:

Perimetro = 2*pi*r**2

# E portanto, tamb√©m possuem a mesma √°rea:

Area = pi*r**2

# Al√©m disso, consideramos que est√£o nas mesmas condi√ß√µes de temperatura (tanto mesma temperatura de base, quanto
# mesma tempertaura do fluido com que ir√£o trocar calor)

Tb = 100
Tamb = 25

M_cobre = (Tb- Tamb)*sqrt(h_*Perimetro*kCu*Area)
M_aluminio = (Tb- Tamb)*sqrt(h_*Perimetro*kAl*Area)

#print(M_cobre)
#print(M_aluminio)

m_cobre = sqrt((h*Perimetro)/(kCu*Area))
m_al = sqrt((h*Perimetro)/(kAl*Area))

# print(m_cobre)
# print(m_al)

num_cobre = (sinh(m_cobre*L_) + (h_/(m_cobre*kCu))*cosh(m_cobre*L_))
den_cobre = (cosh(m_cobre*L_) + (h_/(m_cobre*kCu))*sinh(m_cobre*L_))
qa_cobre = M_cobre*(num_cobre/den_cobre)

print(num_cobre)
print(den_cobre)
# 0. 072

num_al = (sinh(m_al*L_) + (h_/(m_al*kAl))*cosh(m_al*L_))
den_al = (cosh(m_al*L_) + (h_/(m_al*kAl))*sinh(m_al*L_))
qa_al = M_aluminio*(num_al/den_al)
# 0.093 

print(num_al)
print(den_al)

print("Esse √© o qa do cobre {0}".format(qa_cobre))
print("Esse √© o qa do alum√≠nio {0}".format(qa_al))
```

Isso porque, ao mesmo tempo que o resultado do M seja maior para o cobre, o resultado da fra√ß√£o acaba sendo menor. Enquanto o alum√≠nio tem um M menor, mas a sua fra√ß√£o √© maior.
