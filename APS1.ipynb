```python
import numpy as np
from math import *
import matplotlib.pyplot as plt


Ts = 100 # ºC
Tinf = 25 # ºC
rAleta = 12e-3 # m
L = 168e-3 # m
h = 15 # W/(m^2*K) 5 15
k = 154.808 # W/(m*K) 190 154.08
P = 2*pi*rAleta
Atr = pi*(rAleta**2)

# Fazer gráfico da temperatura da aleta por comprimento da aleta
```

1 - a) Ajuste o modelo e defina as condições de contorno para obter um gráfico da distribuição de temperatura (T_analítica) ao longo do comprimento da aleta. Qual a temperatura na extremidade da aleta?

```python
# teta/ tetab = T(x) - Tinf/ Tb - Tinf

# T(x) = 75* (cosh m*(L-x) + (h/m*k)*senh m(L-x))/(cosh m*L + (h/m*k)*senh m*L )

import math
import numpy as np

m = sqrt((h*P)/(k*Atr))
listaTemp = []
lista_x = np.linspace(0,L, 10000)
for x in lista_x:
    num = (cosh(m*(L-x)) + (h/(m*k))*sinh(m*(L-x)))
    den = (cosh(m*L) + (h/(m*k))*sinh(m*L))
    temp = ((num/den)*(Ts - Tinf)) + 25
    listaTemp.append(temp)


plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(lista_x, listaTemp, color="darkcyan")
plt.title("Gráfico de T(x) por x")
plt.xlabel("Valores de x [de 0 até L em metros]")
plt.ylabel("Valores de T(x) em Celsius - caso A")
```

1 - b) Determine a taxa de transferência de calor, a eficiência e a efetividade para a geometria da aleta usada. 

```python
# taxa de transferência de calor
M = (Ts- Tinf)*sqrt(h*P*k*Atr)
qa = M*((np.sinh(m*L) + (h/(m*k))*cosh(m*L))/(cosh(m*L) + (h/(m*k))*np.sinh(m*L)))

areaTotal = Atr+(2*pi*rAleta*L)
taxa_transf_calor = qa
eficiencia = qa/(h*areaTotal*(Ts - Tinf))
efetividade = qa/ (h*Atr*(Ts-Tinf))

print(f'Área total {areaTotal}')
print(f'A taxa de transferência de calor é {taxa_transf_calor}')
print(f'A eficiência é {eficiencia}')
print(f'A efetividade é {efetividade}')

```

* Se um bastão de cobre com o mesmo diâmetro fosse usado em lugar do bastão de alumínio, qual seria a mudança na taxa de remoção de calor através do bastão? Considere kCu= 400 W/(m*K) e 𝑘Al = 240 W/(m*K)



```python
kCu = 400 #W/(mK)
kAl = 240 #W/(mK)
h_ = 5
r = 0.05
L_ = 0.1

# se ambos tem o mesmo diâmetro, também possuem o mesmo raio, que será chamado de r;
# Portanto, o perímetro do bastão circular será o mesmo, já que:

Perimetro = 2*pi*r**2

# E portanto, também possuem a mesma área:

Area = pi*r**2

# Além disso, consideramos que estão nas mesmas condições de temperatura (tanto mesma temperatura de base, quanto
# mesma tempertaura do fluido com que irão trocar calor)

Tb = 100
Tamb = 25

M_cobre = (Tb- Tamb)*sqrt(h_*Perimetro*kCu*Area)
M_aluminio = (Tb- Tamb)*sqrt(h_*Perimetro*kAl*Area)

#print(M_cobre)
#print(M_aluminio)

m_cobre = sqrt((h*Perimetro)/(kCu*Area))
m_al = sqrt((h*Perimetro)/(kAl*Area))

# print(m_cobre)
# print(m_al)

num_cobre = (sinh(m_cobre*L_) + (h_/(m_cobre*kCu))*cosh(m_cobre*L_))
den_cobre = (cosh(m_cobre*L_) + (h_/(m_cobre*kCu))*sinh(m_cobre*L_))
qa_cobre = M_cobre*(num_cobre/den_cobre)

print(num_cobre)
print(den_cobre)
# 0. 072

num_al = (sinh(m_al*L_) + (h_/(m_al*kAl))*cosh(m_al*L_))
den_al = (cosh(m_al*L_) + (h_/(m_al*kAl))*sinh(m_al*L_))
qa_al = M_aluminio*(num_al/den_al)
# 0.093 

print(num_al)
print(den_al)

print("Esse é o qa do cobre {0}".format(qa_cobre))
print("Esse é o qa do alumínio {0}".format(qa_al))
```

Embora o k do cobre seja maior que o k do alumínio, na equação esses números acabam não fazendo muita diferença.
Isso porque, ao mesmo tempo que o resultado do M seja maior para o cobre, o resultado da fração acaba sendo menor. Enquanto o alumínio tem um M menor, mas a sua fração é maior. Assim, há uma pequena diferença e de fato, o cobre apresenta uma taxa de transferência um pouco maior, mas essa diferença acaba não sendo muito grande.
